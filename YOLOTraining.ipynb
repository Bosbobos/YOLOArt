# Обучение YOLO11l на датасете оружия

```python
from roboflow import Roboflow
import os
from ultralytics import YOLO
import torch
from IPython.display import Image, display
```

Для скачивания датасета понадобится api-ключ roboflow. Его нужно закинуть в файл cfg.py в следующем виде: `roboflow_api = '<key>'`

```python
from cfg import roboflow_api
rf = Roboflow(api_key=roboflow_api)
project = rf.workspace("nick-wong-fkzox").project("gun-detector-e4r9p")
dataset = project.version(1).download("yolov5")
```

```python
import logging

loggers = [logging.getLogger(name) for name in logging.root.manager.loggerDict]
for logger in loggers:
    logger.disabled = True
```

```python
model = YOLO('yolo11l.pt',
             verbose=False)

results = model.train(
    data=os.path.join(dataset.location, 'data.yaml'),
    epochs=100,
    imgsz=640,
    batch=16,
    seed=17,
    verbose=False
)
```

```python
model = YOLO('models/best(3).pt',
             verbose=False)
model.val(data=os.path.join(dataset.location, 'data.yaml'),imgsz=640)
```
